```r
N <- 10000000
a <- rnorm(N)
b <- rnorm(N)
```

```r
#vectorized operation
c <- a * b
c
```

```r
#non vectorized operation
d <- rep(NA,N)
for(i in 1:N){
    d[i] <- a[i] * b[i]
    }
d
```
